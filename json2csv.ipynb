```python
import json
import csv
import glob
import pandas as pd
```

```python
json_files = [
'/Users/kruthikhulisandra/Desktop/SpotifyProject/Spotify Extended Streaming History/Streaming_History_Audio_2021-2022_0.json',
'/Users/kruthikhulisandra/Desktop/SpotifyProject/Spotify Extended Streaming History/Streaming_History_Audio_2022-2023_1.json',
'/Users/kruthikhulisandra/Desktop/SpotifyProject/Spotify Extended Streaming History/Streaming_History_Audio_2023-2024_2.json',
'/Users/kruthikhulisandra/Desktop/SpotifyProject/Spotify Extended Streaming History/Streaming_History_Audio_2024-2025_3.json'
]
```

```python
columns = [
    "ts", "platform", "ms_played", "conn_country", "ip_addr",
    "master_metadata_track_name", "master_metadata_album_artist_name",
    "master_metadata_album_album_name", "spotify_track_uri", "episode_name",
    "episode_show_name", "spotify_episode_uri", "audiobook_title",
    "audiobook_uri", "audiobook_chapter_uri", "audiobook_chapter_title",
    "reason_start", "reason_end", "shuffle", "skipped", "offline",
    "offline_timestamp", "incognito_mode"]
```

```python
with open('streaming2125.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=columns)
    writer.writeheader()  # Write the header once

    # Process each JSON file
    for json_file in json_files:
        with open(json_file, 'r') as f:
            data = json.load(f)
            for entry in data:
                # Filter the entry to only include the desired columns
                row = {col: entry.get(col, None) for col in columns}
                writer.writerow(row)
```
