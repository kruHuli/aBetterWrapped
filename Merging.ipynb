```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
dfS= pd.read_csv("dfSCleaned.csv")
```

```python
songInfo = pd.read_csv("merged_songs.csv")
```

```python
songInfo.columns
```

```python
dfS.columns
```

```python
dfM = dfS.merge(songInfo, right_on='Track ID', left_on='spotifyID', how='inner')
dfM
```

```python
dfM.columns
```

```python
final_columns = [
    'date', 'time',                  # Keep if you're exploring temporal patterns
    'track_name', 'artist_name', 'album_name', 'Record Label',
    'spotifyID',
    'ms_played', 'skipped',          # Useful for engagement metrics
    'Duration (ms)', 'Popularity', 'Genres',
    'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness',
    'Instrumentalness', 'Liveness', 'Valence', 'Tempo'
]

dfM = dfM[final_columns]
```

```python
dfM
```

```python
cols_to_average = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness',
                   'Instrumentalness', 'Liveness', 'Valence', 'Tempo']

# Calculate daily averages grouped by 'date' and 'skipped'
daily_avg_skipped = dfM.groupby(['date', 'skipped'])[cols_to_average].mean().reset_index()

# Pivot the averages to separate skipped True/False into individual columns
daily_avg_pivot = daily_avg_skipped.pivot(index='date', columns='skipped', values=cols_to_average)

# Flatten MultiIndex columns for clarity
daily_avg_pivot.columns = [f"{col}_daily_avg_skipped_{str(skip)}" for col, skip in daily_avg_pivot.columns]

# Reset index to merge
daily_avg_pivot = daily_avg_pivot.reset_index()

# Merge these averages back into your original dataframe (dfM)
dfM = dfM.merge(daily_avg_pivot, on='date', how='left')
```

```python
dfM
```

```python
dfM['time'] = pd.to_datetime(dfM['time'])

# Include both hour and minute for precision
dfM['hour_fraction'] = dfM['time'].dt.hour + dfM['time'].dt.minute / 60.0

# Convert to radians (24-hour cycle)
dfM['time_radians'] = 2 * np.pi * dfM['hour_fraction'] / 24.0

# Apply the -1*cos(x) transformation
dfM['time_cos'] = -1 * np.cos(dfM['time_radians'])

# Optional cleanup
dfM.drop(['hour_fraction', 'time_radians'], axis=1, inplace=True)
```
